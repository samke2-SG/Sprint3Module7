```python
plt.bar(last_recorded_data['COUNTRY_NAME'],last_recorded_data['TotalCase'])

plt.xlabel("Contries name")
plt.ylabel("Total Cases")
plt.title("Statistics of Countries Affected by COVID19\n" + "Worldwide,25 APRIL 2020 compared to a year ago")
plt.show()
```

```python
First_3 = last_recorded_data.head(3)
plt.bar(First_3['COUNTRY_NAME'],First_3['TotalCase'], color=['red', 'teal', 'purple', 'green',])

plt.xlabel("Contries name")
plt.ylabel("Total Cases")
plt.title("Statistics of First_3 Countries Affected by COVID19\n" + "Worldwide,25 APRIL 2020 compared to a year ago")

plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.1', color='black')
# Customize the minor grid
plt.grid(which='minor', linestyle=':', linewidth='0.1', color='black')
plt.show()
plt.show()
```

```python
Last_3 = last_recorded_data.tail(3)
plt.bar(Last_3['COUNTRY_NAME'],Last_3['TotalCase'], color=['red', 'teal', 'purple', 'green',])

plt.xlabel("Contries name")
plt.ylabel("Total Cases")
plt.title("Statistics of Last_3 Countries Affected by COVID19\n" + "Worldwide,25 APRIL 2020 compared to a year ago")
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.1', color='black')
# Customize the minor grid
plt.grid(which='minor', linestyle=':', linewidth='0.1', color='black')
plt.show()
plt.show()
```

```python
total_cases = last_recorded_data['TotalCase']
New_cases = last_recorded_data['NewCase']
Death_cases = last_recorded_data['TotalDeath']


plt.scatter(Death_cases, New_cases, label='New Cases', color='teal')
plt.scatter(Death_cases, Death_cases, label='Death cases', color='purple')
plt.title('Scatter Plot')
plt.ylabel('Cases world wide')
plt.xlabel('cases Range')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from pylab import randn
X = Death_cases
Y = total_cases 
plt.scatter(X,Y, color='teal')
plt.xlabel("Death_cases")
plt.ylabel("total_cases ")
plt.show()
```
